```python
import pandas as pd
import numpy as np
from datetime import date
from datetime import datetime
```

```python
 
df = pd.read_table("https://ia800900.us.archive.org/21/items/third-eye/2017-09-07-tweets.tsv", names=["datetime", "network", "duration", "network_info", "text"])
```

```python
lst = pd.date_range(start="2017-09-07",end="2018-09-01", freq="D")
```

```python
d = pd.DataFrame()
```

```python

d["2017-09-07"] = lst
d["2017-09-11"] = pd.to_datetime(d["2017-09-07"]).dt.to_period(freq="D")
```

```python
d["2017-09-11"] = d["2017-09-11"].dt.strftime("%Y-%m-%d")
```

```python
d.head()
```

```python
d.tail()
```

```python
type(d.iloc[1,1])
```

```python
old_url = "https://ia800900.us.archive.org/21/items/third-eye/2017-09-07-tweets.tsv"
old_date = "2017-09-07"
for x in range(1, len(d)):
    
    new_date = d.iloc[x,1]
    print(new_date)
    old_url = "https://ia800900.us.archive.org/21/items/third-eye/2017-09-07-tweets.tsv"
    new_url = old_url.replace(old_date, new_date)
    try:
        df_new = pd.read_table(new_url, names=["datetime", "network", "duration", "network_info", "text"])
        df = pd.concat([df, df_new], axis=0)
    except Exception:
        continue


 


```

```python
df_2017 = df.copy()
```

```python
df_2017.head()
```

```python
df_2017.tail()
```
