To prepare this for concatenation with health data, need to make health LocationName and df ID the same. ID has one extra digit for the block group, while LocationName only includes census tracts. Combining the block groups into census tracts by averaging the features scaled by population.

```python
df['LocationName'] = loc // 10
```

```python
df.head()
```

```python
df['PWDIS'].isnull().value_counts()
```

```python
df[df['PWDIS'] == 0]
```

```python
df['PWDIS'].fillna(0, inplace=True)
```

```python
df.head()
```

```python
# Changing index to LocationName

df.set_index('LocationName', inplace=True)
```

```python
df.head()
```

```python
# Multiplying each column (other than POP) by population

for column in df:
    if column != 'POP':
        df[column] *= df['POP']

df.head()
```

```python
# Grouping rows with same LocationName. Each numerical value is given the population-weighted mean

def mean_normed_by_pop(x):
    x = x.sum() / x['POP'].sum()
    return x

df = df.groupby('LocationName', group_keys=False).apply(mean_normed_by_pop)
```

```python
df.head()
```

```python
# POP was also changed, so it needs to be dropped

df.drop('POP', axis=1, inplace=True)
```

```python
# Re-indexing

df.reset_index(names='LocationName', inplace=True)
```

**Environmental risk columns defined** 

PM25 = Particulate matter 2.5 level in air  

Ozone = Ozone level in air  

DSLPM = Diesel particulate matter level in air  

Cancer = Air toxics cancer risk  

RESP = Air toxics respiratory hazard index  

PTRAF = Traffic proximity and volume  

PRE1960PCT = % pre-1960 housing (lead paint indicator)  

PNPL = Proximity to National Priorities List (NPL) [superfund] sites  

PRMP = Proximity to Risk Management Plan (RMP) facilities, facilities that use extremely hazardous substances  
       
PTSDF = Proximity to Treatment Storage and Disposal facilities (TSDF) 

PWDIS = Indicator for major direct dischargers to water
